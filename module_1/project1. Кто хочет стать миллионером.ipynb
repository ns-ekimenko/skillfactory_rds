# 31. Названия фильмов какой студии в среднем самые длинные по количеству символов?
Варианты ответа:
1. Universal Pictures (Universal)
2. Warner Bros
3. Jim Henson Company, The
4. Paramount Pictures
5. Four By Two Productions

```python
word_count = Counter()
film_count = Counter()
for i,x in df.production_companies.str.split('|').iteritems():
    word_count += Counter({x_i: len(df.loc[i].original_title) for x_i in x})
    film_count += Counter(x)

mean_film_title_length = {}    
for key in word_count:
    mean_film_title_length[key] = word_count[key] / film_count[key]
    
sorted(mean_film_title_length.items(), key=lambda x:x[1], reverse=True)
```
